**Handling Outliers using IQR**

```python
#Removing outliers
# Define a function to remove outliers
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers_iqr(df, 'Quantity')
df = remove_outliers_iqr(df, 'Consumption')
```

**Adding Necessary Columns**

New Columns are added such that to increase the accessibility of the product 

```python
#Finding Ratio od Consumption
df['Consumption_Ratio'] = df['Consumption'] / df['Quantity']
```

```python
#Consumption levels
def categorize_consumption(value):
    if value <= df["Consumption"].quantile(0.33):
        return "Low"
    elif value <= df["Consumption"].quantile(0.66):
        return "Medium"
    else:
        return "High"

df["Consumption_Level"] = df["Consumption"].apply(categorize_consumption)

df['Quantity_consumption_ratios'] = df['Quantity']/df['Consumption']
```

```python
#Product Average quantity
df['Product_Avg_Quantity'] = df.groupby('Product_ID')['Quantity'].transform('mean')
```
